## 2. Cross-Project Integration

```python
from anvil.engine import AnvilEngine, AnvilConfig
from anvil.integrations import (
    VerifyLoopBridge,
    ErrorRecoveryBridge,
    AgentSwarmBridge,
    CostOptimizerBridge,
)

# Set up the core engine
engine = AnvilEngine(
    model_backend="local",
    model_name="fableforge-14b",
    workspace=".",
)

console.print("[bold green]Core engine initialized[/bold green]")
console.print(f"  Model: fableforge-14b")
console.print(f"  Tools: {len(engine.tools)} built-in")
console.print(f"  Verification: enabled")
```

```python
# Wire up all four ecosystem bridges

# 1. VerifyLoop — iterative verification convergence
verify_bridge = VerifyLoopBridge(
    max_rounds=5,
    convergence_threshold=0.95,
)
console.print("[cyan]✓ VerifyLoopBridge[/cyan] — Iterative verify-fix cycles")

# 2. ErrorRecovery — intelligent error healing
recovery_bridge = ErrorRecoveryBridge(
    strategy="cascade",  # retry → rewrite → escalate
    max_attempts=3,
)
console.print("[cyan]✓ ErrorRecoveryBridge[/cyan] — Cascade error recovery")

# 3. AgentSwarm — multi-agent delegation
swarm_bridge = AgentSwarmBridge(
    agents=["coder", "reviewer", "tester"],
    routing="auto",
)
console.print("[cyan]✓ AgentSwarmBridge[/cyan] — Multi-agent delegation")

# 4. CostOptimizer — smart model routing
cost_bridge = CostOptimizerBridge(
    default_model="fableforge-14b",
    budget_usd=5.0,
    strategy="best_value",
)
console.print("[cyan]✓ CostOptimizerBridge[/cyan] — Cost-aware model routing")

# Attach cost optimizer to engine
engine = cost_bridge.attach(engine)
console.print("\n[bold green]All bridges configured![/bold green]")
```

## 7. Full Ecosystem Integration Example

```python
# Put it all together: Anvil + all bridges + bench-agent
# This is the recommended production setup

from anvil.engine import AnvilEngine, AnvilConfig
from anvil.integrations import (
    VerifyLoopBridge,
    ErrorRecoveryBridge,
    CostOptimizerBridge,
)

async def full_pipeline_example():
    """Complete ecosystem integration example."""
    
    # 1. Set up engine with cost optimization
    engine = AnvilEngine(
        config=AnvilConfig(
            model_backend="local",
            model_name="fableforge-14b",
            verify=True,
            max_iterations=15,
            max_retries=3,
            sandbox=True,
            cost_limit_usd=5.0,
        )
    )
    
    # 2. Attach cost optimizer
    cost_bridge = CostOptimizerBridge(
        default_model="fableforge-14b",
        budget_usd=5.0,
        strategy="best_value",
    )
    engine = cost_bridge.attach(engine)
    
    # 3. Set up verification loop
    verify_bridge = VerifyLoopBridge(
        max_rounds=5,
        convergence_threshold=0.95,
    )
    
    # 4. Set up error recovery
    recovery_bridge = ErrorRecoveryBridge(
        strategy="cascade",
        max_attempts=3,
    )
    
    # 5. Run the full pipeline
    task = "Implement user registration with email validation and unit tests"
    
    try:
        # Run with verification loop
        result = await verify_bridge.run(task=task, engine=engine)
        
        console.print(Panel(
            f"[bold green]Task Complete![/bold green]\n"
            f"Summary: {result.summary}\n"
            f"Verification: {result.verification.score:.2f}\n"
            f"Duration: {result.duration_seconds:.1f}s\n"
            f"Cost: ${result.cost.total_usd:.4f}",
            title="Full Pipeline Result",
        ))
        
    except Exception as e:
        # Fall back to error recovery
        result = await recovery_bridge.heal(
            error=e,
            context=result,
            engine=engine,
        )
        console.print(f"[yellow]Recovered from error:[/yellow] {type(e).__name__}")
    
    return result

# Run the example
# result = await full_pipeline_example()
console.print("[dim]Uncomment to run full pipeline (requires running model).[/dim]")
console.print("[bold green]Ecosystem integration configured successfully![/bold green]")
```
